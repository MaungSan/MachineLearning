```python
import urllib2
from bs4 import BeautifulSoup
```

```python
articleURL = "https://www.washingtonpost.com/powerpost/conservative-groups-and-lawmakers-demanding-full-repeal-could-derail-obamacare-rollback/2017/03/02/0bf3f1a0-feaa-11e6-8ebe-6e0dbe4f2bca_story.html?hpid=hp_hp-top-table-main_acaconservatives-1130a%3Ahomepage%2Fstory&utm_term=.5b9a9f423f8e"
```

```python
def getTextFromPage(url):
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = urllib2.Request(url,headers=hdr)
    page = urllib2.urlopen(req).read().decode('utf8','ignore') 
    soup = BeautifulSoup(page,"lxml")
    text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
    return text.encode('ascii',errors='replace').replace("?", " ")
```

```python
text = getTextFromPage(articleURL)
text
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sents =sent_tokenize(text)
sents
```

```python
word_sents = word_tokenize(text.lower())
word_sents
```

```python
_stopwordsEng = stopwords.words('english')
_stopwordsEng
```

```python
_punc =list(punctuation)
_punc
```

```python
_stopwords =set(_stopwordsEng + _punc)
_stopwords
```

```python
word_sent = [word for word in word_sents if word not in _stopwords]
word_sent
```

```python
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
```

```python
from heapq import nlargest
```

```python
nlargest(10, freq, key=freq.get)
```

```python
from collections import defaultdict
ranking = defaultdict(int)
ranking
```

```python
for i, sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
ranking
```

```python
sents_idx = nlargest(3, ranking, key= ranking.get)
sorted(sents_idx)
```

```python
[sents[j] for j in sorted(sents_idx)]
```
